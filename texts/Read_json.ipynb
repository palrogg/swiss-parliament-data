```python
import json
import pandas as pd
```

```python
with open('textes.json', 'r') as f:
    data = json.loads(f.read())
```

```python
df_national = pd.read_csv('../councillors/data/national.csv')
df_etats = pd.read_csv('../councillors/data/states.csv')
```

```python
dfc = pd.concat([df_national, df_etats])
```

```python
len(data.keys())
```

```python
rows = []
for councillor_id, transcripts in data.items():
    for transcript in transcripts:
        rows.append([councillor_id, transcript])
```

```python
df = pd.DataFrame(rows, columns=['councillor_id', 'transcript'])
```

```python
df['councillor_id'] = df['councillor_id'].astype(int)
```

```python
dfc_usecols = ['canton', 'cantonName', 'faction',
       'factionName', 'firstName', 'function', 'id', 'lastName',
       'number', 'party', 'partyName', 'council']
```

```python
dfm = df.merge(dfc[dfc_usecols], left_on='councillor_id', right_on='id', how='left')
```
