# Get Votes

Récupérer les votes des parlementaires actuels en JSON grâce à l’API du parlement

http://ws-old.parlament.ch/

Votes:
http://ws-old.parlament.ch/votes/councillors

Exemple de liste de votes:


http://ws-old.parlament.ch/votes/councillors/3055?legislativePeriodFiter=662

```python
import pandas as pd
import json
import requests
from time import sleep
pd.set_option('display.max_colwidth', 999)
```

```python
# df à actualiser avec Get_Councellors.ipynb
df_national = pd.read_csv('../councillors/data/national.csv')
df_etats = pd.read_csv('../councillors/data/states.csv')
```

```python
df_national['factionName'].value_counts()
```

```python
# Si on veut cibler un groupe précis:
# for i, row in df_national[df_national['factionName'] == 'Groupe PDC'].iterrows():

# On peut aussi filtrer avec legislativePeriodFilter=662

for i, row in df_national.iterrows():
    print(i, row['firstName'], row['lastName'], row['id'], row['number'])
    number = row['id']
    response = requests.get('http://ws-old.parlament.ch/votes/councillors/{}?format=json&lang=fr'.format(number), headers={'Accept': 'application/json', 'Accept-Language': 'fr-CH, fr;q=0.9, en;q=0.8, de;q=0.7, *;q=0.5'})
    try:
        df = pd.DataFrame(response.json())
    except:
        print('JSON invalide pour', row['firstName'], row['lastName'])
        continue
    
    # On enregistre la réponse non aplatie
    print('Shape', df.shape)
    df.to_csv('data/raw/{}.csv'.format(number))
    
    affairs = df['affairVotes'].tolist()
    df_affair = pd.DataFrame(affairs)
    
    # On aplatit et enregistre
    dfm = df.join(df_affair, rsuffix='_')
    del dfm['affairVotes']
    del dfm['councillorVote']
    dfm.to_csv('data/national/{}_votes.csv'.format(number))
    sleep(.5)
```

```python
# Si on veut cibler un groupe précis:
# for i, row in df_national[df_national['factionName'] == 'Groupe PDC'].iterrows():

for i, row in df_etats[7:].iterrows():
    print(i, row['firstName'], row['lastName'], row['id'], row['number'])
    number = row['number']
    response = requests.get('http://ws-old.parlament.ch/votes/councillors/{}?format=json&lang=fr'.format(number), headers={'Accept': 'application/json', 'Accept-Language': 'fr-CH, fr;q=0.9, en;q=0.8, de;q=0.7, *;q=0.5'})
    try:
        df = pd.DataFrame(response.json())
    except:
        print('JSON invalide pour', row['firstName'], row['lastName'])
        continue

    # On enregistre la réponse non aplatie
    print('Shape', df.shape)
    df.to_csv('data/raw/{}.csv'.format(number))
    
    affairs = df['affairVotes'].tolist()
    df_affair = pd.DataFrame(affairs)
    
    # On aplatit et enregistre
    dfm = df.join(df_affair, rsuffix='_')
    del dfm['affairVotes']
    del dfm['councillorVote']
    dfm.to_csv('data/etats/{}_votes.csv'.format(number))
    sleep(.5)
```
