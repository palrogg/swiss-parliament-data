```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/nombre de candidatures et d\'élus depuis 1975.csv', encoding='latin-1', sep=';', skiprows=1)
```

```python
df['Part d’élus'] = df['Elus'] / df['Candidatures'] * 100
```

```python
df.index = df['Année']
```

```python
df[df['Sexe'] == 'Femme']
```

```python
plt.style.use('fivethirtyeight')
```

```python
ax = df.groupby('Sexe')['Elus'].plot(legend=True)
```

```python
df.groupby('Sexe')['Part d’élus'].plot(legend=True)
```

```python
df.groupby('Sexe')['Candidatures'].plot(legend=True)
```
